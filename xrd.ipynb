# Téléchargement du ﬁchier CIF du matériau

```python
from pymatgen.ext.matproj import MPRester

with MPRester("Igrq5N9t2TZdFvec") as m:

    structure = m.get_structure_by_material_id("mp-961706")
```

# 3 premiers pics du diffractogramme correspondant pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator

xrd = XRDCalculator(1.54060, 1)
pattern = xrd.get_pattern(structure)

print("Premier pic :\n")
print("Indices hkl associés : ", pattern.hkls[0][0]["hkl"], "de multiplicité ", pattern.hkls[0][0]["multiplicity"])
print("Position : 2\u0398 = ", pattern.x[0],"°")
print("Intensité : ", pattern.y[0],"% \n")

print("Deuxième pic :\n")
print("Indices hkl associés : ", pattern.hkls[1][0]["hkl"], "de multiplicité ", pattern.hkls[1][0]["multiplicity"])
print("Position : 2\u0398 = ", pattern.x[1],"°")
print("Intensité : ", pattern.y[1],"% \n")

print("Troisième pic :\n")
print("Indices hkl associés : ", pattern.hkls[2][0]["hkl"], "de multiplicité ", pattern.hkls[2][0]["multiplicity"])
print("Position : 2\u0398 = ", pattern.x[2],"°")
print("Intensité : ", pattern.y[2],"% \n")

    
xrd.show_plot(structure, two_theta_range=(0,45))
```
